```python
class DirectedGraphPonderado:
    def __init__(self,gdict=None):
        if gdict is None:
            self.__gdict={}
        else:
            
            self.__gdict={}
            if len(gdict[list(gdict.keys())[0]][0])!=2:
                for V in gdict.keys():
                    self.__gdict[V]=[]
                    for n in range(len(gdict[V])):                       

                        self.__gdict[V].append([gdict[V][n],1])
            else:
                self.__gdict=gdict
            
                     
        
    def addVertex(self,vertex):
        if vertex not in self.__gdict.keys():
            self.__gdict[vertex]=[]
            return True
        return False
        
    def addEdge(self,vertex1,vertex2, peso=1):
        if vertex1 in self.__gdict.keys() and vertex2 in self.__gdict.keys():
            self.__gdict[vertex1].append([vertex2,peso])
            
            return True
        return False
    
    def remove_Edge(self,vertex1,vertex2):
        if vertex1 in self.__gdict.keys() and vertex2 in self.__gdict.keys():
            try:
                self.__gdict[vertex1].remove([vertex2,self.getPeso(vertex1,vertex2)])
                
            except ValueError:
                pass
            return True
        return False
    
    def getPeso(self,vertex1,vertex2):
        count=0
        pos=0
        for V in self.__gdict[vertex1]:
            if V[0]==vertex2:
                pos=count
                break
            count=count+1
        return self.__gdict[vertex1][pos][1]
    
    def UpdatePeso(self,vertex1,vertex2,p):
        if vertex1 in self.__gdict.keys() and vertex2 in self.__gdict.keys():
            count=0
            pos=0
            for V in self.__gdict[vertex1]:
                if V[0]==vertex2:
                    pos=count
                    break
                count=count+1
            self.__gdict[vertex1][pos][1]=p
        
            return True
        return False
    
    def remove_Vertex(self,vertex):
        if vertex in self.__gdict.keys():
            other_vertex_list=[]
            for V1 in self.__gdict.keys():
                for V2 in self.__gdict[V1]:
                    if V2[0] is vertex:
                        other_vertex_list.append(V1)
                        
            for other_vertex in other_vertex_list:
                self.__gdict[other_vertex].remove([vertex,self.getPeso(other_vertex,vertex)])
                
            del self.__gdict[vertex]
            return True
        return False
        
    def printGraph(self):
        for vertex in self.__gdict:
            print(vertex,":",self.__gdict[vertex])
    
    def Degree(self,vertex):
        if vertex in self.__gdict.keys():
            return len(self.__gdict[vertex])
        return -1
            
    def __Degree_per_Vertex(self):
        self.__DegreeDict={}
        for V in self.__gdict.keys():
            self.__DegreeDict[V]= self.Degree(V)
        return self.__DegreeDict
    
    def number_of_vertex(self):
        return len(self.__gdict.keys())
    
    def number_of_Edges(self):
        Dg= self.__Degree_per_Vertex()
        Sg=0
        for V in Dg.keys():
            Sg=Sg+Dg[V]
        return Sg
    
    def Dijkstra(self, inicio):
        if inicio not in self.__gdict:
            return "O vertice nao existe no grafo."

        d = {vertex: float('inf') for vertex in self.__gdict}
        d[inicio] = 0
        N = set()
        P = {}

        while True:
    
            min_distance = float('inf')
            min_vertex = None
            for vertex in self.__gdict:
                if vertex not in N and d[vertex] < min_distance:
                    min_distance = d[vertex]
                    min_vertex = vertex

            if min_vertex is None:# Condicao de Parada
                break

            N.add(min_vertex)

            for neighbor, C in self.__gdict[min_vertex]:
                dist = d[min_vertex] + C

                if dist < d[neighbor]:
                    d[neighbor] = dist
                    P[neighbor] = min_vertex

        return d, P



    def shortest_path(self, start, end):
        D, pred = self.Dijkstra(start)

        path = []
        current_vertex = end

        while current_vertex != start:
            path.append(current_vertex)
            current_vertex = pred.get(current_vertex)
            if current_vertex is None:
                return "Nenhum caminho encontrado entre os vertices."

        path.append(start)
        path.reverse()
        return path
```
